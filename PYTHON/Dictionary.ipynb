```python
#1. Create a Dictionary (Student ID → Name)
students = {
    101: "Amit",
    102: "Riya",
    103: "Kiran",
    104: "Sita",
    105: "Rahul"
}

print("Original Dictionary:", students)
```

```python
#1.1 Adding values in dictionary
students[106] = "Vikram"
print("After Adding:", students)
```

```python
#1.2 Updating values in dictionary
students[102] = "Riya Sharma"
print("After Updating:", students)
```

```python
#1.3 Accessing value in dictionary
print("Name of ID 103:", students[103])
```

```python
#1.4 Create a nested dictionary
nested_students = {
    101: {"name": "Amit", "age": 20},
    102: {"name": "Riya", "age": 21},
    103: {"name": "Kiran", "age": 19}
}

print("Nested Dictionary:", nested_students)
```

```python
#1.5 Access values of nested dictionary
print("Student 101 Name:", nested_students[101]["name"])
print("Student 102 Age:", nested_students[102]["age"])
```

```python
#1.6 Print keys present in a dictionary
print("Keys:", students.keys())
```

```python
#.7 Delete a value from dictionary
del students[104]
print("After Deletion:", students)
```
