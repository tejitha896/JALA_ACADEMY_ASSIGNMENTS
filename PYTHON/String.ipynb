```python
#1. Different ways creating a string
# Using single quotes
s1 = 'Hello'

# Using double quotes
s2 = "World"

# Using triple quotes (multi-line)
s3 = """This is a multi-line string"""
print(s1)
print(s2)
print(s3)
```

```python
#2. Concatenating two strings using + operator
s1 = input("enter first string : ")
s2 = input("enter second string : ")

result = s1 + " " + s2
print(result)  # Hello World
```

```python
#3. Finding the length of the string
s=input("enter a string : ")
print(len(s))
```

```python
#4. Extract a string using Substring
s="this is python program"
print(s[0:3])
print(s[2:8])
```

```python
#5. Searching in strings using index()
# Take input from user
s = input("Enter a string: ")
sub = input("Enter substring to find: ")

# Check if substring exists before using index()
if sub in s:
    print("Index:", s.index(sub))
else:
    print("Substring not found")
```

```python
#6. Matching a String Against a Regular Expression With matches()
import re

s = "Hello123"

# Check if string starts with letters and ends with numbers
if re.match(r"[A-Za-z]+\d+", s):
    print("Match found")
```

```python
#7. Comparing strings
s1 = input("enter a string : ")
s2 = input("enter a string : ")

print(s1 == s2)   # False
print(s1 < s2)    # True (lexicographical comparison)
```

```python
#8. startsWith(), endsWith() and compareTo()
# Take input from user
s = input("Enter a string: ")

# Check startswith
start = input("Enter starting text to check: ")
print("Starts with:", s.startswith(start))

# Check endswith
end = input("Enter ending text to check: ")
print("Ends with:", s.endswith(end))

# Take two strings for comparison
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

# Compare strings (equivalent to compareTo)
result = (s1 > s2) - (s1 < s2)

print("Comparison result:", result)
```

```python
#9. Trimming strings with strip()
s = "   Hello   "
print(s)
print(s.strip())
```

```python
#10. Replacing characters in strings with replace()
# Take input from user
s = input("Enter a string: ")
# Take word to be replaced
old = input("Enter the word to replace: ")
# Take new word
new = input("Enter the new word: ")
# Replace the word
result = s.replace(old, new)
# Print result
print("Updated string:", result)
```

```python
#11. Splitting strings with split()
text = input("Enter words separated by spaces: ")
words = text.split()

print("List of words:", words)
```

```python
#12. Converting integer objects to Strings
num = int(input("Enter a number: "))
text = str(num)

print("Number as string:", text)
print("Type:", type(text))
```

```python
#13. Converting to uppercase and lowercase
text = input("Enter a string: ")

print("Uppercase:", text.upper())
print("Lowercase:", text.lower())
```
