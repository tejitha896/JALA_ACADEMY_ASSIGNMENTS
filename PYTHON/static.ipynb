```python
#1. Define a static variable and access that through a class 
class Example:
    # Static (class) variable
    static_var = 10

# Accessing static variable using the class name
print("Access using class:", Example.static_var)
```

```python
#2. Define a static variable and access that through a instance
class Example:
    # Static (class) variable
    static_var = 100

# Create an instance (object) of the class
obj = Example()

# Access the static variable using the instance
print("Access using instance:", obj.static_var)
```

```python
#3. Define a static variable and change within the instance
class Example:
    # Static (class) variable
    static_var = 100

# Create an instance (object)
obj = Example()

# Access before change
print("Before change:", Example.static_var)  # 100

# Change the static variable using the instance
obj.static_var = 200

# Access after change
print("After change using instance:", obj.static_var)      # 200
print("Access using class:", Example.static_var)           # Still 100
```

```python
#4. Define a static variable and change within the class
class Example:
    # Static (class) variable
    static_var = 100

    # Method to change the static variable
    def change_value():
        Example.static_var = 200  # Changing using class name

# Access before change
print("Before change:", Example.static_var)  # 100

# Call the method to change the value
Example.change_value()

# Access after change
print("After change:", Example.static_var)   # 200
```
