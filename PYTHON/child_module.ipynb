```python
from main_module import ParentClass

class ChildClass(ParentClass):
    def access(self):
        # Accessing protected members from parent class
        print("Accessing from child class:")
        print("Number:", self._number)
        self._protected_method()


if __name__ == "__main__":
    obj = ChildClass()
    obj.access()
```

```python
from main_module import ParentClass

class ExternalClass:
    def access(self):
        obj = ParentClass()

        # Accessing public fields
        print("Number:", obj.number)
        print("Message:", obj.message)

        # Accessing public method
        obj.public_method()


if __name__ == "__main__":
    obj = ExternalClass()
    obj.access()
```
