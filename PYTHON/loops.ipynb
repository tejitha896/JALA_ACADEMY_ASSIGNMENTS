```python
#8. Write a program to find Armstrong number or not
def armstrong(n):
    temp = n #stores the value because the initial changes during the process and need to compare it with total value
    total = 0

    while temp > 0:
        digit = temp % 10  #used for taking single digit from combined number
        total = total + digit**3   # fixed power
        temp = temp // 10  #used for remove the one's digit number

    if total == n:
        print("Armstrong number")
    else:
        print("Not Armstrong")
num=int(input("enter a number : "))
armstrong(num)
```

```python
#9. Write a program to find the prime or not.
num=int(input("enter your number : "))
count=0
for i in range (1,num+1):
    if num%i==0:
        count=count+1
if count>2:
    print("Not prime")
else:
    print("Prime number")
```

```python
#10. Write a program to palindrome or not.
num=int(input("enter a number : "))
rev=0
temp=num
while num>0:
    rev=rev*10+num%10
    num=num//10

if rev==temp:
    print("Palindrom number")
else :
    print("Not palindrom")
```

```python
#11. Program to check whether a number is EVEN or ODD using switch
num = int(input("Enter a number: "))

def even(n):
    if n % 2 == 0:
        print("Even number")
    else:
        print("Not even")

def odd(n):
    if n % 2 != 0:
        print("Odd number")
    else:
        print("Not odd")

ch = int(input("Enter choice (1 for Even, 2 for Odd): "))

switch = {
    1: even,
    2: odd
}

switch[ch](num)
```
