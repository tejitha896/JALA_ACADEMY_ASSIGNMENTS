```python
#1. Write a program to read text file
with open("input.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#2.Write text to a .txt file using InputStream-like method
import io
data = io.BytesIO(b"Hello, this is written using a stream.")

with open("output.txt", "wb") as file:
    file.write(data.read())
```

```python
#3. Read a file stream
with open("input.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#4. Read a file with random access
with open("input.txt", "r") as file:
    file.seek(15)  # move to byte position 15
    print(file.read())
```

```python
#5. Read from a particular index using seek()
with open("input.txt", "r") as file:
    file.seek(10)  # start from index 10
    data = file.read()
    print(data)
```

```python
#6. Check read/write permissions of a file
import os

file_path = "input.txt"

if os.path.exists(file_path):
    print("Readable:", os.access(file_path, os.R_OK))
    print("Writable:", os.access(file_path, os.W_OK))
else:
    print("File does not exist.")
```
