```python
#1. Write a program to generate Arithmetic Exception without exception handling
# This will raise ZeroDivisionError (Arithmetic Exception)
a = 10
b = 0
# No try-except block, so program crashes
result = a / b

print(result)
```

```python
#2. Handle the Arithmetic exception using try-catch block
try:
    a = 10
    b = 0
    result = a / b  # This will raise exception
except ZeroDivisionError:
    print("Arithmetic Exception handled: Division by zero is not allowed")
```

```python
#3. Write a method which throws exception, Call that method in main class without try block
# Method that raises exception
def divide(a, b):
    return a / b

# Calling without try-except
result = divide(10, 0)

print(result)
```

```python
#4. Write a program with multiple catch blocks
try:
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    result = a / b
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input, please enter integers")
except Exception as e:
    print("General exception:", e)
```

```python
#5. Write a program to throw exception with your own message
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")

print("Age is:", age)
```

```python
#6. Write a program to create your own exception
# Custom exception class
class MyException(Exception):
    pass

try:
    raise MyException("This is a custom exception")
except MyException as e:
    print("Caught custom exception:", e)
```

```python
#7. Write a program with finally block
try:
    a = 10
    b = 0
    result = a / b
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("This block always executes")
```
