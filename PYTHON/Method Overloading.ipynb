```python
#1. Write two methods with the same name but different number of parameters of same type and call the methods
class Example:
    # Method with one or two parameters (same type: int)
    def show(self, a, b=None):
        # If only one parameter is passed
        if b is None:
            print("One parameter:", a)
        else:
            # If two parameters are passed
            print("Two parameters:", a, b)

# Creating object
obj = Example()

# Calling method with one parameter
obj.show(10)

# Calling method with two parameters
obj.show(10, 20)
```

```python
#2. Write two methods with the same name but different number of parameters of different data type and call the methods 
class Example:
    # Method that handles different data types using type checking
    def show(self, a, b=None):
        # If only one parameter is given
        if b is None:
            print("One parameter:", a)
        else:
            # Check type of second parameter
            if isinstance(b, str):
                print("Second parameter is string:", a, b)
            else:
                print("Second parameter is not string:", a, b)

# Creating object
obj = Example()

# Calling with one parameter
obj.show(10)

# Calling with integer
obj.show(10, 20)

# Calling with string
obj.show(10, "Hello")
```

```python
#3. Write two methods with the same name and same number of parameters of same type 
class Example:
    # First method definition
    def show(self, a):
        print("First method:", a)

    # Second method with same name and parameters
    def show(self, a):
        print("Second method:", a)

# Creating object
obj = Example()

# This will call only the second method (it overrides the first)
obj.show(10)
```
