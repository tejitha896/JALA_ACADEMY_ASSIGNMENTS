```python

#3. Define variables for different Data Types int, Boolean, char, float, double and print on the Console

num=int(input("enter number in (int or float or double) :"))#take the value in int or float or double
print("The user number is : ",num,"and the datatype is : ",type(num))#takes the input from user and print datatype

flag=True
print("Boolean Value :",flag, "and data type is :" ,type(flag))#takes the input from user and print datatype

ch = input("Enter a character: ")# takes the sring or char from user

if len(ch) == 1: #check the string value length and return the related statement
    print("Valid character:", ch,"and the type is :",type(ch))
else:
    print("Please enter only one character")
```

```python
#4.Define the local and Global variables with the same name and print both variables and understand the scope of the variable
x = 10   # Global variable
def show():
    x = 20   # Local variable (same name as global)
    print("Inside function (Local x):", x)

show()

print("Outside function (Global x):", x)
```
