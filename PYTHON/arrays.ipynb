```python
#1. Write a function to add integer values of an array
def sum_array(arr):
    total = 0
    for num in arr:
        total += num
    return total
arr=list(map(int,input().split()))
print(sum_array(arr))  # Output: 10
```

```python
#2. Write a function to calculate the average value of an array of integers
def average_array(arr):
    total = 0
    count = 0

    for num in arr:
        total += num
        count += 1

    return total / count

# Example
arr = list(map(int, input("Enter numbers: ").split()))
print("Average:", average_array(arr))  # Output: 2.5
```

```python
#3. Write a program to find the index of an array element
arr = list(map(int, input("Enter elements: ").split()))
key = int(input("Enter element to find: "))

def find_index(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

print("Index:", find_index(arr, key))
```

```python
#7. Write a function to insert an element at a specific position in the array
def insert_element(arr, pos, value):
    new_arr = []
    n = len(arr)

    for i in range(n + 1):
        if i < pos:
            new_arr.append(arr[i])
        elif i == pos:
            new_arr.append(value)
        else:
            new_arr.append(arr[i - 1])

    return new_arr

# User input
arr = list(map(int, input("Enter elements: ").split()))
pos = int(input("Enter position (0-based index): "))
value = int(input("Enter value to insert: "))

result = insert_element(arr, pos, value)

print("Updated array:", result)
```

```python
#8. Write a function to find the minimum and maximum value of an array
def find_min_max(arr):
    # Check if the array is empty
    if len(arr) == 0:
        return None, None  # No minimum or maximum exists

    # Assume the first element is both min and max initially
    minimum = arr[0]
    maximum = arr[0]

    # Loop through each element in the array
    for num in arr:
        # If current number is smaller than minimum, update minimum
        if num < minimum:
            minimum = num

        # If current number is greater than maximum, update maximum
        if num > maximum:
            maximum = num

    # Return both minimum and maximum values
    return minimum, maximum

arr = list(map(int, input("Enter elements: ").split()))

min_val, max_val = find_min_max(arr)

print("Minimum:", min_val)  
print("Maximum:", max_val)  
```

```python
#9. Write a function to reverse an array of integer values
def reverse_array(arr):
    reversed_arr = []  # Create an empty list to store reversed elements

    # Loop through the array from the last element to the first
    for i in range(len(arr) - 1, -1, -1):
        reversed_arr.append(arr[i])  # Add each element to the new list

    return reversed_arr  # Return the reversed array

arr = list(map(int, input("Enter elements: ").split()))

print("Reversed Array:", reverse_array(arr))  # Output: [5, 4, 3, 2, 1]
```

```python
#10. Write a function to find the duplicate values of an array
def find_duplicates(arr):
    duplicates = []  # List to store duplicate values

    # Loop through each element in the array
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            # Check if two elements are the same
            if arr[i] == arr[j]:
                # Add to duplicates list if not already added
                if arr[i] not in duplicates:
                    duplicates.append(arr[i])

    return duplicates  # Return all duplicate values

arr = list(map(int, input("Enter elements: ").split()))

print("Duplicate values:", find_duplicates(arr))  # Output: [1, 2]
```

```python
#11. Write a program to find the common values between two arrays
def find_common_values(arr1, arr2):
    common = []  # List to store common elements

    # Loop through first array
    for i in range(len(arr1)):
        # Loop through second array
        for j in range(len(arr2)):
            # Check if elements are the same
            if arr1[i] == arr2[j]:
                # Add to common list if not already added
                if arr1[i] not in common:
                    common.append(arr1[i])

    return common  # Return common elements


arr1= list(map(int, input("Enter elements: ").split()))
arr2= list(map(int, input("Enter elements: ").split()))

print("Common values:", find_common_values(arr1, arr2))  # Output: [3, 4, 5]
```

```python
#12. Write a method to remove duplicate elements from an array
def remove_duplicates(arr):
    result = []  # List to store unique elements

    # Loop through each element in the array
    for num in arr:
        # Add the element only if it is not already in the result list
        if num not in result:
            result.append(num)

    return result  # Return array without duplicates

arr= list(map(int, input("Enter elements: ").split()))
print("Array after removing duplicates:", remove_duplicates(arr))  
```

```python
#13. Write a method to find the second largest number in an array
def second_largest(arr):
    # Check if the array has at least two elements
    if len(arr) < 2:
        return None  # Not enough elements

    # Initialize first and second largest
    first = second = float('-inf')

    # Loop through each number in the array
    for num in arr:
        # If current number is greater than the largest
        if num > first:
            second = first   # Update second largest
            first = num      # Update largest

        # If number is between first and second largest
        elif first > num > second:
            second = num

    return second  # Return the second largest number

arr= list(map(int, input("Enter elements: ").split()))
print("Second largest number:", second_largest(arr))  
```

```python
#15. Write a method to find number of even number and odd numbers in an array
def count_even_odd(arr):
    even_count = 0  # Counter for even numbers
    odd_count = 0   # Counter for odd numbers

    # Loop through each number in the array
    for num in arr:
        # Check if number is even
        if num % 2 == 0:
            even_count += 1
        else:
            # Otherwise, it is odd
            odd_count += 1

    return even_count, odd_count  # Return both counts

arr= list(map(int, input("Enter elements: ").split()))
even, odd = count_even_odd(arr)

print("Even numbers:", even)  
print("Odd numbers:", odd)    
```

```python
#16. Write a function to get the difference of largest and smallest value
def difference_max_min(arr):
    # Check if the array is empty
    if len(arr) == 0:
        return None  # No values to compare

    # Initialize smallest and largest with the first element
    smallest = arr[0]
    largest = arr[0]

    # Loop through each element in the array
    for num in arr:
        # Update smallest if a smaller value is found
        if num < smallest:
            smallest = num

        # Update largest if a larger value is found
        if num > largest:
            largest = num

    # Return the difference between largest and smallest
    return largest - smallest

arr= list(map(int, input("Enter elements: ").split()))
print("Difference:", difference_max_min(arr))  # Output: 15 (20 - 5)
```
