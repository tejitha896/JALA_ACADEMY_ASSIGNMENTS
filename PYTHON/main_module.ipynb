****1. Create a class with PRIVATE fields, private method and a main method. Print the fields
in main method. Call the private method in main method.
Create a sub class and try to access the private fields and methods from sub class****

```python
class ParentClass:
    def __init__(self):
        self.__number = 10
        self.__message = "Hello from private field"

    # Private method
    def __private_method(self):
        print("This is a private method.")

    def access_within_class(self):
        print("Accessing private data within class:")
        print("Number:", self.__number)
        print("Message:", self.__message)
        self.__private_method()


class ChildClass(ParentClass):
    def try_access(self):
        # These lines will cause errors if uncommented
        # because private members are not accessible

        # print(self.__number)         # AttributeError
        # print(self.__message)        # AttributeError
        # self.__private_method()      # AttributeError

        print("Cannot access private members from subclass.")


# Main execution
if __name__ == "__main__":
    obj = ParentClass()

    # Access private members within the same class
    obj.access_within_class()

    print("\nTrying to access from subclass:")
    child = ChildClass()
    child.try_access()
```

****2. Create a class with PROTECTED fields and methods. Access these fields and methods
from any other class in the same package.
Also, Access the PROTECTED fields and methods from child class located in a different
package
Access the PROTECTED fields and methods from any class in different package****

```python
class ParentClass:
    def __init__(self):
        self._number = 100   # Protected field

    def _protected_method(self):  # Protected method
        print("This is a protected method")


class AnotherClass:
    def access(self):
        obj = ParentClass()

        # Accessing protected members (allowed, but not recommended)
        print("Number:", obj._number)
        obj._protected_method()


if __name__ == "__main__":
    obj = AnotherClass()
    obj.access()
```

```python
class ParentClass:
    def __init__(self):
        self.number = 50   # Public field
        self.message = "Hello from public field"

    def public_method(self):  # Public method
        print("This is a public method")


class AnotherClass:
    def access(self):
        obj = ParentClass()

        # Accessing public fields
        print("Number:", obj.number)
        print("Message:", obj.message)

        # Accessing public method
        obj.public_method()


if __name__ == "__main__":
    obj = AnotherClass()
    obj.access()
```
