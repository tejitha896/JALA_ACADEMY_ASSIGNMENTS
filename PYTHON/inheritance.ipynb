*****7. Inheritance
A, B and C are classes
A is a super class. B is a sub class of A. C is a sub class of B.
Create three methods in each class, 2 methods are specific to each class and third
method (override method) should be in all three Classes A, B and C
Create a class with main method. Create an object for each class A, B and C in main
method and call every method of each class using its own object/instance.
Call an overridden method with super class reference to B and C class’s objects
Runtime Polymorphism with Data Members/Instance variables, Repeat the above
process only for data members*****

```python
# Superclass A
class A:
    def __init__(self):
        self.var_a = "A's variable"

    def method_a1(self):
        print("Method A1")

    def method_a2(self):
        print("Method A2")

    def display(self):  # overridden method
        print("Display from Class A")


# Subclass B
class B(A):
    def __init__(self):
        super().__init__()
        self.var_b = "B's variable"

    def method_b1(self):
        print("Method B1")

    def method_b2(self):
        print("Method B2")

    def display(self):  # override
        print("Display from Class B")


# Subclass C
class C(B):
    def __init__(self):
        super().__init__()
        self.var_c = "C's variable"

    def method_c1(self):
        print("Method C1")

    def method_c2(self):
        print("Method C2")

    def display(self):  # override
        print("Display from Class C")
```

```python
def main():
    # Object of A
    obj_a = A()
    obj_a.method_a1()
    obj_a.method_a2()
    obj_a.display()
    print()

    # Object of B
    obj_b = B()
    obj_b.method_a1()
    obj_b.method_a2()
    obj_b.method_b1()
    obj_b.method_b2()
    obj_b.display()
    print()

    # Object of C
    obj_c = C()
    obj_c.method_a1()
    obj_c.method_a2()
    obj_c.method_b1()
    obj_c.method_b2()
    obj_c.method_c1()
    obj_c.method_c2()
    obj_c.display()
    print()

    # Runtime Polymorphism
    ref = obj_b  # A type reference to B object
    ref.display()  # calls B's display

    ref = obj_c  # A type reference to C object
    ref.display()  # calls C's display


if __name__ == "__main__":
    main()
```

```python
class A:
    def __init__(self):
        self.x = 10

    def show(self):
        print("Class A method")


class B(A):
    def __init__(self):
        super().__init__()
        self.x = 20

    def show(self):
        print("Class B method")


class C(B):
    def __init__(self):
        super().__init__()
        self.x = 30

    def show(self):
        print("Class C method")
```

```python
def main_vars():
    obj_b = B()
    obj_c = C()

    # Reference to B object
    ref = obj_b
    ref.show()
    print("x =", ref.x)

    # Reference to C object
    ref = obj_c
    ref.show()
    print("x =", ref.x)


if __name__ == "__main__":
    main_vars()
```
