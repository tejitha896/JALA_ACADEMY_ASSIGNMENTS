```python
#1. Write a class with a default constructor, one argument constructor and two argument constructors. Instantiate the class to call all the constructors of that class from a main class
class Student:
    def __init__(self, name=None, age=None):
        if name is None and age is None:
            print("Default constructor called")
            self.name = "Unknown"
            self.age = 0

        elif age is None:
            print("One-argument constructor called")
            self.name = name
            self.age = 0

        else:
            print("Two-argument constructor called")
            self.name = name
            self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")


# Main code
if __name__ == "__main__":
    # Default constructor
    s1 = Student()
    s1.display()

    # One-argument constructor
    s2 = Student("Alice")
    s2.display()

    # Two-argument constructor
    s3 = Student("Bob", 20)
    s3.display()
```

```python
#2. Call the constructors(both default and argument constructors) of super class from a child class
class Parent:
    # Default constructor
    def __init__(self, name="Default Parent"):
        self.name = name
        print("Parent constructor called")
        print("Parent name:", self.name)


class Child(Parent):
    # Child constructor
    def __init__(self, name, age):
        # Calling parent constructor (argument constructor)
        super().__init__(name)
        self.age = age
        print("Child constructor called")
        print("Child age:", self.age)


# Main code
if __name__ == "__main__":
    print("Calling child with argument:")
    c1 = Child("John", 10)

    print("\nCalling parent directly (default behavior):")
    p1 = Parent()
```

```python
#3. Apply private, public, protected and default access modifiers to the constructor
class Demo:
    # Public constructor (default in Python)
    def __init__(self, name):
        self.name = name
        print("Public constructor called")
        print("Name:", self.name)

    # Protected-like constructor (by convention)
    def _protected_constructor(self, value):
        print("Protected constructor called")
        print("Value:", value)

    # Private-like constructor (name mangling)
    def __private_constructor(self, data):
        print("Private constructor called")
        print("Data:", data)

    # Method to access private constructor internally
    def call_private(self):
        self.__private_constructor("Secret Data")


# Main code
if __name__ == "__main__":
    # Public constructor
    obj1 = Demo("Public Name")

    # Calling protected constructor (possible but discouraged)
    obj1._protected_constructor(100)

    # Calling private constructor via a public method
    obj1.call_private()
```

```python
#4. Write a program which illustrates the concept of attributes of a constructor
class Student:
    # Constructor with attributes
    def __init__(self, name, age, course):
        # Attributes of the constructor
        self.name = name
        self.age = age
        self.course = course

        print("Constructor called and attributes initialized")

    # Method to display attributes
    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Course:", self.course)


# Main program
if __name__ == "__main__":
    s1 = Student("Alice", 20, "Computer Science")
    s1.display()

    print()

    s2 = Student("Bob", 22, "Mathematics")
    s2.display()
```
