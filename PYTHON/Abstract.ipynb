```python
#1. Create an abstract class with abstract and non-abstract methods.
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    # Abstract method (must be implemented by subclasses)
    @abstractmethod
    def make_sound(self):
        pass

    # Non-abstract method (has implementation)
    def sleep(self):
        print("Animal is sleeping")

# Subclass implementing the abstract method
class Dog(Animal):

    def make_sound(self):
        print("Dog barks")

# Main code
d = Dog()
d.make_sound()  # Abstract method implementation
d.sleep()       # Inherited non-abstract method
```

```python
#2. Create a sub class for an abstract class. Create an object in the child class for the abstract class and access the non-abstract methods
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    
    # Abstract method
    @abstractmethod
    def make_sound(self):
        pass
    
    # Non-abstract method
    def sleep(self):
        print("Animal is sleeping")

# Subclass (Child class)
class Dog(Animal):
    
    # Implementing the abstract method
    def make_sound(self):
        print("Dog barks")

# Creating object of subclass
d = Dog()

# Calling methods
d.make_sound()  # Abstract method implementation
d.sleep()       # Non-abstract method from abstract class
```

```python
#3. Create an instance for the child class in child class and call abstract methods
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass

# Child class
class Dog(Animal):
    
    # Implementing abstract method
    def make_sound(self):
        print("Dog barks")
    
    # Method inside child class
    def create_and_call(self):
        # Creating instance of child class inside the same class
        obj = Dog()
        
        # Calling abstract method using the object
        obj.make_sound()

# Creating object of child class
d = Dog()

# Calling method that creates another instance
d.create_and_call()
```

```python
#4. Create an instance for the child class in child class and call non-abstract methods
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass
    
    # Non-abstract method
    def sleep(self):
        print("Animal is sleeping")

# Child class
class Dog(Animal):
    
    # Implementing abstract method
    def make_sound(self):
        print("Dog barks")
    
    # Method inside child class
    def create_and_call(self):
        # Creating instance of child class inside itself
        obj = Dog()
        
        # Calling non-abstract method from abstract class
        obj.sleep()

# Creating object of child class
d = Dog()

# Calling method
d.create_and_call()
```
