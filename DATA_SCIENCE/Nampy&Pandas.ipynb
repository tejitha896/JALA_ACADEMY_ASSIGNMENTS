**1.Numpy**

```python
# Q1: Create a NumPy array and normalize using min-max scaling

import numpy as np  # Import NumPy

arr = np.random.randint(1, 100, 10)  
# Create array with random integers

min_val = np.min(arr)  
# Find minimum value

max_val = np.max(arr)  
# Find maximum value

normalized = (arr - min_val) / (max_val - min_val)  
# Apply min-max scaling formula

print(normalized)  # Display result
```

```python
# Q2: Replace specific value in 5x5 matrix

import numpy as np

matrix = np.random.randint(1, 10, (5,5))  
# Create 5x5 matrix

matrix[matrix == 5] = 99  
# Replace all 5 with 99

print(matrix)
```

```python
# Q3: Element-wise operations

import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division
```

```python
# Q4: Solve linear equations using NumPy

import numpy as np

A = np.array([[3,4],[5,2]])  
# Coefficient matrix

B = np.array([7,8])  
# Constants

solution = np.linalg.solve(A, B)  
# Solve equations

print(solution)
```

```python
# Q5: Broadcasting operation

import numpy as np

matrix = np.ones((3,3))  
# Create 3x3 matrix

arr = np.array([1,2,3])  
# Create 1D array

result = matrix + arr  
# Broadcasting addition

print(result)
```

```python
# Q7: Matrix multiplication

import numpy as np

a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

result = np.dot(a, b)  
# Matrix multiplication

print(result)
```

```python
# Q8: Dot product and cross product

import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))  # Dot product
print(np.cross(a,b))  # Cross product
```

```python
# Q9: Find unique elements

import numpy as np

arr = np.random.randint(1, 10, 20)  
# Create array

unique = np.unique(arr)  
# Find unique values

print(unique)
```

```python
# Q10: Function to find inverse of matrix

import numpy as np

def inverse_matrix(mat):  
    # Define function
    return np.linalg.inv(mat)  
    # Return inverse

matrix = np.array([[1,2],[3,4]])

print(inverse_matrix(matrix))
```

**Data Cleaning & Preprocessing**

```python
# Q21: Clean non-numeric values from dataset

df = pd.read_csv("data.csv")  # Load dataset

df['col1'] = pd.to_numeric(df['col1'], errors='coerce')  
# Convert to numeric, invalid values → NaN

df.dropna(inplace=True)  
# Remove rows with invalid data

print(df)
```

```python
# Q22: Remove duplicate rows

df = pd.read_csv("data.csv")  # Load dataset

df.drop_duplicates(inplace=True)  
# Remove duplicate rows

print(df)
```

```python
# Q23: Normalize using StandardScaler

from sklearn.preprocessing import StandardScaler  # Import scaler

scaler = StandardScaler()  # Create scaler object

df['scaled'] = scaler.fit_transform(df[['value']])  
# Standard scaling

print(df)
```

```python
# Q24: Label encoding categorical column

from sklearn.preprocessing import LabelEncoder  # Import encoder

le = LabelEncoder()  # Create encoder

df['category_encoded'] = le.fit_transform(df['category'])  
# Convert category to numeric

print(df)
```

```python
# Q25: Train-test split (80-20)

from sklearn.model_selection import train_test_split  # Import split

X = df[['col1','col2']]  # Features
y = df['value']  # Target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=True)  
# Split dataset

print(X_train.shape, X_test.shape)
```
