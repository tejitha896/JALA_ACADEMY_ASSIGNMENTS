1. Data Import and Cleaning

```python
# Q1: Import a CSV file into a Pandas DataFrame and drop rows with missing values

import pandas as pd  # Import pandas library

df = pd.read_csv("data.csv")  # Load CSV file into DataFrame

df_clean = df.dropna()  # Remove rows that contain any missing values

print(df_clean)  # Display cleaned DataFrame
```

```python
# Q2: Replace missing numerical values with the mean of the column

import pandas as pd  # Import pandas
import numpy as np  # Import NumPy

df = pd.read_csv("data.csv")  # Load dataset

df.fillna(df.mean(numeric_only=True), inplace=True)  
# Replace missing values in numeric columns with mean

print(df)  # Display updated DataFrame
```

```python
# Q3: Replace missing categorical values with mode

import pandas as pd  # Import pandas

df = pd.read_csv("data.csv")  # Load dataset

for col in df.select_dtypes(include='object'):  
    # Loop through categorical columns
    
    df[col].fillna(df[col].mode()[0], inplace=True)  
    # Replace missing values with most frequent value (mode)

print(df)  # Display result
```

3. Merging and Joining

```python
# Q7: Merge two DataFrames and fill missing values

import pandas as pd  # Import pandas

df1 = pd.DataFrame({'id':[1,2], 'val':[10,20]})  # First DataFrame
df2 = pd.DataFrame({'id':[2,3], 'val2':[30,40]})  # Second DataFrame

merged = pd.merge(df1, df2, on='id', how='outer')  
# Merge on 'id' using outer join

merged.fillna(0, inplace=True)  
# Replace missing values with 0

print(merged)  # Display result
```

```python
# Q8: Perform left join and handle missing data

result = pd.merge(df1, df2, how='left', on='id')  
# Left join

result.fillna("Missing", inplace=True)  
# Fill missing values

print(result)  # Show result
```

```python
# Q9: Concatenate two DataFrames and handle duplicate column names

concat = pd.concat([df1, df2], axis=1)  
# Combine DataFrames column-wise

concat.columns = ['id1','val1','id2','val2']  
# Rename columns to avoid duplicates

print(concat)  # Display result
```
